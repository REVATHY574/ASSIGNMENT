```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (Replace this with your actual data)
data = {
    'gender': ['male', 'female', 'female', 'male', 'female'],
    'parental_education': ['bachelor', 'master', 'associate', 'bachelor', 'associate'],
    'math_score': [80, 90, 85, 75, 95],
    'reading_score': [85, 88, 92, 78, 96],
    'writing_score': [88, 92, 90, 82, 98],
    'test_preparation_course': ['completed', 'none', 'none', 'completed', 'completed']
}

df = pd.DataFrame(data)

# a. Count of males and females
gender_counts = df['gender'].value_counts()
print("Number of males and females:\n", gender_counts)

# b. Parental level of education
plt.figure(figsize=(10, 6))
sns.countplot(x='parental_education', data=df, order=df['parental_education'].value_counts().index)
plt.title('Parental Level of Education')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()

# c. Average scores based on gender and test preparation course
average_scores_gender = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()
average_scores_course = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean()

print("Average scores based on gender:\n", average_scores_gender)
print("\nAverage scores based on test preparation course:\n", average_scores_course)

# d. Scoring variation based on gender and test preparation course
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='gender', y='math_score', data=df)
plt.title('Math Score Variation based on Gender')

plt.subplot(1, 2, 2)
sns.boxplot(x='test_preparation_course', y='math_score', data=df)
plt.title('Math Score Variation based on Test Preparation Course')

plt.show()

# e. Bonus points for the top 25% of students based on math score
top_25_percent = df[df['math_score'] >= df['math_score'].quantile(0.75)]
print("Top 25% of students based on math score:\n", top_25_percent)
```
